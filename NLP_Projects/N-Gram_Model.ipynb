### Task #2

<u>Process the downloaded <code>csv</code> files in python.</u>

There's a nice csv library already included in python for accessing values in that are stored in a comma separated values (csv) format. Read the <a href="https://docs.python.org/3/library/csv.html">csv library documentation</a>.
What is the delimiter in your csv files? Open each of the two .csv files that you downloaded using this library and be able to read in the data. Note that we really only care about the text column in this assignment.

```python
# Reading the wine-reviews dataset as a csv file and keeping only the text
# column in a list
wine_reviews_list = [] # this list will contain the values of the text column
with open('winemag-data-130k-v2.csv', newline='',encoding="utf8") as wine_reviews_f:
    wine_reviews_csv = csv.reader(wine_reviews_f, delimiter=',', quotechar='"')
    # Skipping the header that contains column names
    next(wine_reviews_csv, None)
    for row in wine_reviews_csv:
        # We choose item with index 2 because the column of index 2 in the
        # dataset is the text column that we want
        wine_reviews_list.append(row[2])
print('There are {} items of wine reviews'.format(len(wine_reviews_list)))
```

```python
# To make later computations faster, we will keep the first 20,000 reviews
wine_reviews_list = wine_reviews_list[:20_000]
```

<u>Perform sentence segmentation and word tokenization.</u>

Utilize the nltk module to perform sentence segmentation and word tokenization. But at this point, there are a few decisions that need to be made:

* How we should handle the .csv rows in the previous step? If we ignore row makers, and "lump everything together", how will that effect our language model?
* Do we want to remove punctuation? What is the effect of keeping punctuation in the model?
* Do we want to add sentence boundary markers, such as <samp>&lt;S&gt;</samp> and <samp>&lt;/S&gt;</samp>?</li>
* Should two the words <samp>The</samp> and <samp>the</samp> be treated as the same? What are the effects of doing, or not doing, this?

```python
# Making everything lowercase in wine reviews and news datasets
wine_reviews_list = [x.lower() for x in wine_reviews_list]
news_list = [x.lower() for x in news_list]
```

```python
# Sentence segmentation and word tokenization
from nltk.tokenize import sent_tokenize, word_tokenize
# For wine reviews
wine_reviews_sentences = [sent_tokenize(x) for x in wine_reviews_list]
wine_reviews_sentences = [item for sublist in wine_reviews_sentences for item in sublist]
# Now the wine_reviews_sentences list is a list of sentences
wine_reviews_words = word_tokenize(' '.join(wine_reviews_sentences))
# For the news
news_sentences = [sent_tokenize(x) for x in news_list]
news_sentences = [item for sublist in news_sentences for item in sublist]
news_words = word_tokenize(' '.join(news_sentences))
# We didn't add sentence boundaries (i.e. <S> and </S>) and kept the punctuation.
# This way, we can now from punctuations when the sentence starts and when it ends
# without the need for sentence boundaries
```

### Task #4

<u>Calculate N-gram counts and compute probabilities.</u>

Use a python dictionary (or any suitable data structure) to first compute unigram counts. Then try bigram counts. Finally, trigram counts.

How much memory are you using? How fast, or slow, is the code -- how long is this step taking? If it is taking too long, try only using a fraction of your corpus: instead of loading the entire .csv file, try only reading the first 1000 rows of data.

Using those counts, compute the probabilities for the unigrams, bigrams, and trigrams, and store those in a new python dictionary (or some other data structure).

```python
from nltk import bigrams, trigrams
from collections import Counter

# For wine reviews
wine_unigrams = wine_reviews_words.copy()
wine_unigrams_count = Counter(wine_unigrams)
wine_bigrams = bigrams(wine_reviews_words)
wine_bigrams_count = Counter(wine_bigrams)
wine_trigrams = trigrams(wine_reviews_words)
wine_trigrams_count = Counter(wine_trigrams)

# For the news
news_unigrams = news_words.copy()
news_unigrams_count = Counter(news_unigrams)
news_bigrams = bigrams(news_words)
news_bigrams_count = Counter(news_bigrams)
news_trigrams = trigrams(news_words)
news_trigrams_count = Counter(news_trigrams)
```

```python
# To compute the probabilities, we will use a nltk functions
# named FreqDist, ConditionalFreqDist, and ConditionalProbDist
from nltk import FreqDist, ConditionalFreqDist, ConditionalProbDist, ELEProbDist
# For wine-reviews dataset

# First for unigrams
wine_fdist_uni = FreqDist(wine_unigrams)
wine_pdist_uni = {}
wine_uni_sum = sum(wine_fdist_uni.values())
for k,v in wine_fdist_uni.items():
    wine_pdist_uni[k] = v / wine_uni_sum
# Then for bigrams
wine_cfdist_bi = ConditionalFreqDist(bigrams(wine_reviews_words))
wine_cpdist_bi = ConditionalProbDist(wine_cfdist_bi, ELEProbDist)
# Then for trigrams
wine_condition_pairs = (((w0, w1), w2) for w0, w1, w2 in trigrams(wine_reviews_words))
wine_cfdist_tri = ConditionalFreqDist(wine_condition_pairs)
wine_cpdist_tri = ConditionalProbDist(wine_cfdist_tri, ELEProbDist)

# For news dataset
# First for unigrams
news_fdist_uni = FreqDist(news_unigrams)
news_pdist_uni = {}
news_uni_sum = sum(news_fdist_uni.values())
for k,v in news_fdist_uni.items():
    news_pdist_uni[k] = v / news_uni_sum
# Then for bigrams
news_cfdist_bi = ConditionalFreqDist(bigrams(news_words))
news_cpdist_bi = ConditionalProbDist(news_cfdist_bi, ELEProbDist)
# Then for trigrams
news_condition_pairs = (((w0, w1), w2) for w0, w1, w2 in trigrams(news_words))
news_cfdist_tri = ConditionalFreqDist(news_condition_pairs)
news_cpdist_tri = ConditionalProbDist(news_cfdist_tri, ELEProbDist)
```

### Task #6

<u>Generate random sentences from the N-grams models for both datasets.</u>
                        
We briefly talked about this idea in class. It's also introduced at a high-level in J&M 4.3. How can a random number in the range [0,1] probabilistically generate a word using your model?

```python
# For wine reviews
# First we choose a random word to start the sentence
import random
print('Examples of randomly generated wine-reviews sentences (based on a probability distribution): ')
print('-'*60)
for i in range(3):
    print(i+1, ') ', sep='', end='')
    first_word = None
    for k,v in wine_pdist_uni.items():
        if round(v, 2) == round(random.random(), 2):
            first_word = k
            break
    print(first_word[0].upper() + first_word[1:], end=' ')
    # Now we generate next words: each word based on the word before it.
    # The function generate() returns a word based on its probability.
    # So if the word "good" comes after the word "really" with a probability
    # of 0.5, then generate() will return "good" after "really" 50% of the time
    current_word = None
    prev_word = first_word
    while current_word != '.':
        current_word = wine_cpdist_bi[prev_word].generate()
        prev_word = current_word
        print(current_word, end=' ')
    print()
```

```python
# For news, we repeat the same process
print('Examples of randomly generated news sentences (based on a probability distribution): ')
print('-'*60)
for i in range(3):
    print(i+1, ') ', sep='', end='')
    first_word = None
    for k,v in news_pdist_uni.items():
        if round(v, 2) == round(random.random(), 2):
            first_word = k
            break
    print(first_word[0].upper() + first_word[1:], end=' ')
    current_word = None
    prev_word = first_word
    while current_word != '.':
        current_word = news_cpdist_bi[prev_word].generate()
        prev_word = current_word
        print(current_word, end=' ')
    print()
```
